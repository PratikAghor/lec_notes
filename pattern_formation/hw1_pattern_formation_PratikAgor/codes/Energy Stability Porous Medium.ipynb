```python
one_comp_porous_med(m, k, Ra_T) = Ra_T * k^2 - (m^2*pi^2+k^2)^2 # derived in class 
two_comp_porous_med(m, k, Ra_T, Ra_S, tau) = 4*tau*(m^2*pi^2+k^2)^4 + 4*k^2*(m^2*pi^2+k^2)^2*(Ra_S - tau*Ra_T) \
- (k^4)*(Ra_S + Ra_T)^2
```

```python
# plot marginal curve for Lambda = 0 
Ra_S = 1e-2
tau = 1e-1
m = 1

P1 = implicit_plot(one_comp_porous_med(m, k, Ra_T), (k, 0, 20), (Ra_T, 0, 100), \
                   linewidth = 3, color = 'red', axes_labels=['$k$', '$Ra_{T}$'], legend_label = 'one comp')
P2 = implicit_plot(two_comp_porous_med(m, k, Ra_T, Ra_S, tau), (k, 0, 20), (Ra_T, 0, 100), \
                   linewidth = 3, color = 'blue', axes_labels=['$k$', '$Ra_{T}$'], legend_label = 'two comps')

P = P1 + P2
```
